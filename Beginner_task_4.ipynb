***IMPORT LIBRARY***

```python
import cv2 as cv
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plot
from skimage import io
from google.colab.patches import cv2_imshow
```

**READING THE IMAGE** 

```python
#image is loaded using imread() function in cv class
image=cv.imread("tobey.jpg")
```

**DISPLAYING THE IMAGE PRIOR TO CONVERSION**

```python
cv2_imshow(image)
```

`Since this is a jpg format image it must be decoded back to rgb format`










```python
#CONVERSION TO RGB FORMAT
rgb_img=cv.cvtColor(image,cv.COLOR_BGR2RGB)
plot.imshow(rgb_img)
```

**CONVERSION TO GREY SCALE IMAGE** 

```python
grey_image=cv.cvtColor(rgb_img,cv.COLOR_BGR2RGB)

plot.imshow(grey_image)
grey_image
```

**INVERTING THE GREY SCALE IMAGE**

```python
invert=cv.bitwise_not(grey_image)

plot.imshow(inverted_image)
plot.imshow(invert)
#image converted to invert version
```

**BLUR IMAGE CONVERSION**

```python
blurred=cv.GaussianBlur(inverted_image,(21,21),0)
plot.imshow(blurred)
#inverted image is blurred
```

**INVERSION OF BLURRED IMAGE**

```python
inverted_blur_img=255 - blurred
plot.imshow(inverted_blur_img)
```

***FINAL SKETCH***

```python
#using the combination of grey image and inevrted blur image for pencil sketch
pencil_sketch=cv.divide(grey_image,inverted_blur_img,scale=255)
plot.imshow(pencil_sketch)
```

```python
#displaying the final sketch
cv2_imshow(pencil_sketch)
```
